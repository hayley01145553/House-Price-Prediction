### stats.chi2_contingency Returns 

chi2 : float, The test statistic.

p : float, The p-value of the test

dof : int , Degrees of freedom

expected : ndarray, same shape as observed , The expected frequencies, based on the marginal sums of the table.


```python
#### Categorical Value  

df1_part_col = df1_part1_categorical_features.columns

for i in range(len(df1_part_col)):
    X = df1_part1_num_features.SalePrice
    Y = df1_part1_categorical_features[df1_part_col[i]]
    result = stats.chi2_contingency(pd.crosstab(X,Y))
    print("*"*30)
    print("SalesPrice and '",df1_part_col[i], "' corr ")
    print("P value : " , result[1])
    print("correlation True/False : ", result[1] <= 0.05)
```
